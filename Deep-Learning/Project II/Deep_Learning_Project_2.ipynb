```python
def normalize_resize(image, label, desired_size=(32, 32)):
    image = tf.cast(image, tf.float32)
    image = tf.divide(image, 255)
    image = tf.image.resize(image, desired_size)
    return image, label

def standardize_resize(image, label, desired_size=(32, 32)):
    image = tf.cast(image, tf.float32)
    image = tf.divide(image, 255)
    image = tf.image.resize(image, desired_size)
    image = tf.image.per_image_standardization(image)
    return image, label

img_height = 32
img_width = 32
train = train_ds.map(standardize_resize)
test = test_ds.map(standardize_resize)
```

```python
def count_occurences(dataset):
    counts = {}
    for _, label in dataset:
        if label.numpy() not in counts:
            counts[label.numpy()] = 1
        else:
            counts[label.numpy()] += 1
    return counts

counts = count_occurences(dataset) 
print(counts)

# take classes that have between 400 and 600 images
classes = [key for key, value in counts.items() if value >= 30 and value <= 40]
```

```python
@tf.function
def filter_classes(image, label, allowed_classes):
    return tf.reduce_any(tf.equal(label, allowed_classes))
```

```python
from sklearn.model_selection import train_test_split

def split_dataset(dataset, allowed_classes, test_size=0.2):

    dataset = dataset.filter(lambda image, label: filter_classes(image, label, allowed_classes))
    numpy_images = []
    numpy_labels = []

    for image, label in dataset:
        numpy_images.append(image.numpy())
        numpy_labels.append(label.numpy())

    numpy_images = np.array(numpy_images)
    numpy_labels = np.array(numpy_labels)
    
    train_split, test_split = train_test_split(list(zip(numpy_images, numpy_labels)), test_size=test_size, random_state=23)

    train_images, train_labels = zip(*train_split)
    test_images, test_labels = zip(*test_split)

    name_mapping = {}
    new_class_names = []
    for i, label in enumerate(classes):
        name_mapping[label] = i
        new_class_names.append(class_names[label])

    train_labels = np.vectorize(name_mapping.get)(train_labels)
    test_labels = np.vectorize(name_mapping.get)(test_labels)

    return np.array(train_images), np.array(train_labels), np.array(test_images), np.array(test_labels), new_class_names

X_train, y_train, X_test, y_test, new_class_names = split_dataset(dataset, classes)

```

**4) Define the model (5%)**

**Note:** you should probably try different settings and choose the *best* one. Below you will find a proposed definition of a network:

- Activation functions - *ReLU*
- At least 3 *Convolutional blocks* - (Conv2D, Activation, BatchNormalization, Dropout, MaxPooling2D)
  - Conv2D - *kernel=(3,3)*, *padding='same'*
  - MaxPooling2D - *pool_size=(2,2)*
- Flatten layer
- At least 2 layers Fully-Connected (Dense)
- Output layer - Dense with number_of_classes outputs (remember to use softmax)

**Note:** You can add *Activation* as a separate layer or as *activation='relu'* parameter in Conv2D

**Step verification:**

Compile the model with *'adam'* optimizer, the Categorical Crossentropy as the loss function, and measure the accuracy value.

```python


conv_block = lambda filters: [
    layers.Conv2D(filters, (3, 3), padding='same'),
    layers.Activation('relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(pool_size=(2, 2)),
]

conv_block_dropout = lambda filters: [
    layers.Conv2D(filters, (3, 3), padding='same'),
    layers.Activation('relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.2),
    layers.MaxPooling2D(pool_size=(2, 2)),
]

data_augmentation = keras.Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(img_height,
                                  img_width,
                                  3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)

num_classes = len(new_class_names)

model = Sequential([
    layers.Conv2D(32, (2, 2), input_shape=(img_height, img_width, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),

    *conv_block(32),
    *conv_block(64),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(num_classes)
])
```

```python
# compile with adam optimizer
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
    )
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, accuracy_score
import seaborn as sns

def plot_learning_curves(history, epochs):
    fig = plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='train')
    plt.plot(history.history['val_accuracy'], label='test')
    plt.legend()
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Accuracy')
    plt.show()

def plot_confusion_matrix(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred = np.argmax(y_pred, axis=1)
    y_test = np.argmax(y_test, axis=1)
    cm = confusion_matrix(y_test, y_pred)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(10, 10))
    sns.heatmap(cm, annot=True, square=True, cmap=plt.cm.Blues, xticklabels=new_class_names, yticklabels=new_class_names)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

def make_stats(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred = np.argmax(y_pred, axis=1)
    y_test = np.argmax(y_test, axis=1)
    
    print('Accuracy:', accuracy_score(y_test, y_pred))
    print('Recall:', recall_score(y_test, y_pred, average='macro'))
    print('Precision:', precision_score(y_test, y_pred, average='macro'))
```

```python
plot_learning_curves(history, epochs)

plot_confusion_matrix(model, X_test, y_test)
```
